```python
# 이미지 크기 알아내는 코드


from PIL import Image
import os

folder_path = '/content/bing_size_disgust'

# 폴더 내의 모든 파일에 대해 크기 확인
for filename in os.listdir(folder_path):
    if filename.endswith('.jpg') or filename.endswith('.png'):  # jpg, png 파일일 경우에만 실행
        file_path = os.path.join(folder_path, filename)
        with Image.open(file_path) as img:
            width, height = img.size
            print(f"{filename}: width={width}, height={height}")
```

```python
# bing_crawl 이미지 크기가 가지각색이기 때문에 크기를 맞추어주는 코드

folder_path = '/content/bing_disgust_data'
new_folder ="/content/bing_size_disgust"

new_width = 800
new_height =600

for filename in os.listdir(folder_path):
    if filename.endswith('.jpg') or filename.endswith('.png'):  # jpg, png 파일일 경우에만 실행
        file_path = os.path.join(folder_path, filename)
        with Image.open(file_path) as img:
            new_img = img.resize((new_width, new_height))
            new_file_path = os.path.join(new_folder, f"{os.path.splitext(filename)[0]}_{new_width}x{new_height}.jpg")
            new_img.save(new_file_path)


```

```python
# 기존 색상 이미지를 흑백 이미지로 바꾸어주는 코드



# 이미지가 있는 폴더 경로
folder_path = '/content/bing_size_disgust'
new_folder = "/content/bing_black_disgust"

# 폴더 내의 모든 파일에 대해 그레이스케일로 변경
for filename in os.listdir(folder_path):
    if filename.endswith('.jpg') or filename.endswith('.png'):  # jpg, png 파일일 경우에만 실행
        file_path = os.path.join(folder_path, filename)
        with Image.open(file_path) as img:
            gray_img = img.convert('L')
            new_file_path = os.path.join(new_folder, f"{os.path.splitext(filename)[0]}_gray.jpg")
            gray_img.save(new_file_path)
```

```python
# 좌우변환 증강하는 기법

from PIL import Image
import os

original_folder = "/content/disgust"
new_folder ="/content/2_disgust"

for filename in os.listdir(original_folder):

  filepath = os.path.join(original_folder , filename)

  if os.path.isfile(filepath) and filename.lower().endswith(('.png','.jpg','.jpeg')):
    image = Image.open(filepath)

    rotated_image = image.transpose(method = Image.FLIP_LEFT_RIGHT)

    new_filepath = os.path.join(new_folder , f"{os.path.splitext(filename)[0]}_2.jpg")

    rotated_image.save(new_filepath)


print("이미지 회전 및 저장 완료")
```
